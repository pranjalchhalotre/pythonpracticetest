```python
def topten():
    n=1
    while n<=10:
        sq =n*n
        yield sq
        n+=1
values = topten()
print(values)
for i in values:
    print(i)
```

```python
def decorator_function(original_function):
    def wrapper():
        print("before function execution")
        original_function()
        print("after function execution")
    return wrapper
@decorator_function
def say_hello():
    print("hello")
say_hello
```

```python
def div(a,b):
    print(a/b)
def smart_div(func):
    def inner(a,b):
        if a<b:
            a,b =b,a
        return func(a,b)
    return inner
div1 =smart_div(div)
div(2,8)
```

```python
class Student:
    def __init__(self,phy,chem,math):
        self.phy=phy
        self.chem =chem
        self.math =math
    @property
    def percentage(self):
        return str((self.phy +self.chem +self.math)/3)+"%"
stu1 = Student(98,97,98)
print(stu1.percentage)
stu1.phy =86
stu1.chem=100
print(stu1.percentage)
```
