```python
#5. Create a static method to check if a number is even.
class NumberCheck:

    @staticmethod
    def is_even(num):
        if num % 2 == 0:
            return "Even Number"
        else:
            return "Odd Number"

print(NumberCheck.is_even(10))
print(NumberCheck.is_even(7))
```

```python
#6. Override __str__() to print object details.
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Student Name: {self.name}, Age: {self.age}"

s1 = Student("Rahul", 20)
print(s1)
```

```python
#7. Delete an object attribute using del.
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

c1 = Car("Toyota", 2000000)

del c1.price   # deleting attribute

print(c1.brand)
```
