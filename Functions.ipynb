```python
#take random number 5 times and find even and odd and print their counts that how many are even and how many are odd.

import random

def count_even_odd():
    even_count = 0
    odd_count = 0

    for i in range(5):
        num = random.randint(1, 100)
        print("Number:", num)

        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    print("Total Even numbers:", even_count)
    print("Total Odd numbers:", odd_count)

count_even_odd()
```

```python

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mult(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b



a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")


if op == "+":
    print("Addition is:", add(a, b))
elif op == "-":
    print("Subtraction is:", sub(a, b))
elif op == "*":
    print("Multiplication is:", mult(a, b))
elif op == "/":
    print("Division is:", divide(a, b))
else:
    print("Invalid operator")
```
