```python
#3. Write a function find_max(*args) that returns the largest value among all the arguments.
def find_max(*args):
    largest_value = args[0]
    for num in args:
        if num > largest_value:
            largest_value = num
    return largest_value

print(find_max(3, 7, 2, 9, 5))
        
```

```python
#6.#  Write a function greet_user(**kwargs) that prints "Hello <name>" if the key name is provided. If name is not provided, print "Hello Guest".
def greet_user(**kwargs):
    if "name" in kwargs:
        print("Hello", kwargs["name"])
    else:
        print("Hello Guest")

greet_user(name="Pranjal")
```

```python
#7.# Write a function print_details(**kwargs) that prints all key-value pairs in the format: key : value
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

print_details(name="Pranjal", age=22, city="harda")
```
