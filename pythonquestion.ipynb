```python
#1. Find the length of the longest substring without repeating characters in a given string
s= input("enter string:")
longest_len =0
for i in range(len(s)):
    sub=" "
    for j in range (i,len(s)):
        if s[j] not in sub:
            sub +=s[j]
        else:
            break
    if len(sub)>longest_len:
        longest_len = len(sub)
print("longest substring length:",longest_len)
```

```python
#2. Group a list of words into anagrams.
words = ["eat","tea","tan","ate"]
groups={}
for i in words:
    key =" ".join(sorted(i))
    if key not in groups:
        groups[key]=[]
    groups[key].append(i)
print(list(groups.values()))
    
```

```python
16. Remove duplicates from a list using set logic while preserving the original order.
def remove_duplicates(lst):
    seen = set()
    result = []

    for i in lst:
        if i not in seen:
            seen.add(i)
            result.append(i)

    return result


l = [1,2,3,2,4,1,5]
print(remove_duplicates(l))
```

```python
17. Invert a dictionary where duplicate values are handled by storing keys in a list.
def invert_dict(d):
    result = {}

    for key, value in d.items():
        if value not in result:
            result[value] = [key]
        else:
            result[value].append(key)

    return result


d = {'a':1, 'b':2, 'c':1, 'd':2, 'e':3}
print(invert_dict(d))
```

```python
19. Sort a dictionary by its values in descending order without directly using sorted() on the dictionary.
d = {'a':3, 'b':1, 'c':5, 'd':2}

items = list(d.items())

for i in range(len(items)):
    for j in range(i+1, len(items)):
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]

sorted_dict = dict(items)

print(sorted_dict)
```
