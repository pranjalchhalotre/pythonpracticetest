```python
# 1. Create a class BankAccount with private attributes balance and account_number. Provide methods to deposit and withdraw money with validation.
class BankAccount:
    def __init__(self,balance,account_number):
        self.__balance = int(balance)
        self.__account_number = account_number
    def deposit(self,amount):
        if amount>0:
            self.__balance += amount
            print(f"Deposited:{amount}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if amount <=0:
            print("Invalid withdrawal amount!")
        elif amount >self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -=amount
            print(f"Withdrawn:{amount}")

    def get_balance(self):
        return self.__balance
    def display(self):
        print(f"account number:{self.__account_number}")
        print(f"balance:{self.__balance}")
acc = BankAccount("12345", 1000)

acc.deposit(500)
acc.withdraw(300)
acc.withdraw(2000)

acc.display()     
```

```python
# 3.Create a class Employee with a private salary attribute. Add a method to increase salary based on performance rating.
class Employee:
    def __init__(self,name,sal):
        self.name = name
        self.__sal =sal
    def increase_sal(self,rating):
        if rating == "good":
            self.__sal *=1.20
        elif rating =="average":
            self.__sal = 1.10
        else:
            print("no increment ")
            return
        print(f"Salary updated based on {rating} performance")
    def get_sal(self):
        return self.__sal
    def display(self):
        print(f"name:{self.name}")
        print(f"salary:{self.__sal}")

     
        
     
        
emp1 = Employee("Pranjal", 50000)

emp1.display()

emp1.increase_sal("good")

emp1.display()    
```

```python
# 4.Design a class PasswordManager where the password is private and can only be changed after verifying the old password
class PasswordManager:
    def __init__(self,password):
        self.__password =password
    def get_password(self):
        return self.__password
        
    def set_password(self,old_password,new_password):
        if old_password == self.__password:
            self.__password = new_password
            print("password updated !")
        else:
            print("Incorrect password!")
    def display(self):
        
        print(f"updated password:{self.__password}")
emp = PasswordManager("pranjal@123")
print(emp.get_password())
emp.set_password("pranjal@123","123@pranjal")
emp.display()
           
    
```

```python
# 5.Create a class Car with a private attribute speed. Restrict speed to a maximum limit using setter methods.
class Car:
    def __init__(self,speed =0):
        self.__speed =speed
    def get_speed(self):
        return self.__speed
        

    def set_speed(self,speed):
        MAX_SPEED =200
        if speed < 0:
            print("Speed cannot be negative!")
        elif speed > MAX_SPEED:
            print(f"Speed cannot exceed {MAX_SPEED}")
            self.__speed = MAX_SPEED   # limit applied
        else:
            self.__speed = speed
            print(f"Speed set to {speed}")

    def display(self):
        print(f"Current speed: {self.__speed}")
car1 = Car()

car1.set_speed(150)   
car1.display()

car1.set_speed(250) 
car1.display()

car1.set_speed(-10)        
```
