```python
#1. Check whether input number is even or odd. Handle invalid input using exception handling.
try:
    num = int(input("enter num:"))

    if num%2==0:
        print("even")
    else:
        print("odd")
except Exception as e:
    print("invalid input : enter numbers only")

    

```

```python
#2. Take a number from user. If input is valid: (using math build-in function)
"""Print ceil if number > 0
Print floor if number < 0
Handle invalid input"""
import math
try:
    num = float(input("enter num:"))
    if num>0:
        print("ceil value",math.ceil(num))
    elif num<0:
        
        print("floor value",math.floor(num))
    else:
        print("number is zero")

except Exception as e:
    print("invalid input")

```

```python
#3. Write a program to divide two numbers entered by user. Handle the case where the denominator is zero. Print the result if no error is found.
try:
    num1 = float(input("enter num:"))
    num2 = float(input("enter num:"))
    result = num1/num2 
    print("Result:",result)
except ZeroDivisionError:
    print("cannot divide by zero")
except ValueError:
    print("Invalid input. Please enter numbers only")
        
    
```
